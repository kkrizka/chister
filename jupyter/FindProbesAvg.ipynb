```python
probes='probes_*.png'
thecut=40
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import scipy.optimize
from math import *
import glob
```

```python
img=np.zeros((480,640),dtype=np.int32)
n=0
print(img.shape)
for imgpath in glob.glob(probes):
    img1=cv2.imread(imgpath,cv2.IMREAD_GRAYSCALE)
    img+=img1
    n+=1
img=img/n

print(img1.shape,img.shape)
print(n)
    
plt.imshow(img/n, cmap='gray')
```

```python
plt.hist(img.flatten(),bins=100);
```

```python
ret,imgpass=cv2.threshold(img,thecut,255,cv2.THRESH_BINARY_INV)
plt.imshow(imgpass, cmap='gray')
imgpass=np.uint8(imgpass) #[200:450,400:600])
cv2.imwrite('probestemplate.png',imgpass)
```
